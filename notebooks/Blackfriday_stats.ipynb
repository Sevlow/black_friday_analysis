```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import tensorflow as tf
%matplotlib inline
```

```python
data = pd.read_csv('../datasets/BlackFriday.csv')
data = data.fillna(value=0)
```

```python
sns.distplot(data['Purchase'],kde=0)
```

```python
for i in data.Age.unique():
    sns.distplot(data['Purchase'][data.Age==i], kde=1,label='{}'.format(i))
    plt.legend()
```

```python
mean_age_purchase = {}
for i in data.Age.unique():
    mean_age_purchase[i] = data.loc[:,'Purchase'][data['Age']==i].mean()
print(mean_age_purchase)
    
```

```python

df = pd.DataFrame(columns=['moyenne_achat','age'])
print(df)
i=0
for key,value in mean_age_purchase.items():
    df.loc[i]=[value,key]
    i += 1
print(df)
print(df.moyenne_achat.std())
sns.set_context('notebook', font_scale=1, rc={'lineswidth':2.5})
sns.barplot(y=df.moyenne_achat, x=df.age)
```
